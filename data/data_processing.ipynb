```python
import os
import pandas as pd
```

```python
path = os.path.join(os.getcwd(), 'raw_data.csv')
df = pd.read_csv(path)
df
```

```python
df.PermissionGroupName = df.PermissionGroupName.str.title()
df.PermissionNameDescription = df.PermissionNameDescription.str.title()
df.ResourceName = df.ResourceName.str.title()
df.AccessLevelName = df.AccessLevelName.str.title()
df.CompanyName = df.CompanyName.str.title()
df
```
